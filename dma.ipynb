### (1) K-Means clustering

```python
from sklearn.cluster import KMeans
from IPython.display import HTML

N = 5

# split data in N folders
def split_data(data):
    h = {}
    df = pd.DataFrame(columns=['BEBIDAS','ALIMENTACAO','MERCEARIA','CONGELADOS','ACOUGUE','HORTIFRUTEGRANJEIRO','BAZAR','LIMPEZA','RESFRIADOS','PERFUMARIA','CLUSTER_5'])
    for i in range(0,N):
        h[i] = data.loc[data['CLUSTER_5'] == i]
#         data = np.array(['a','b','c','d'])
        s = pd.Series(h[i].mean())
        df = df.append(s, ignore_index = True)
#     display(HTML(df.to_html()))
#         h[i].drop(['MERCEARIA','CLUSTER'],axis=1).plot.hist(bins=9, alpha=0.5)
    return h,df

# get percentage of elements in column
def get_percentage(data,col):
    print (data[col].value_counts(normalize=True) * 100)

# run k-means
def kmeans_func(data, num_clust):
    model = KMeans(num_clust)
    model.fit(data)
    clust_labels = model.predict(data)
    cent = model.cluster_centers_
    return (clust_labels, cent)

clust_labels, cent = kmeans_func(X.drop(['CPF'],axis=1), N)
kmeans = pd.DataFrame(clust_labels)
# del X['CLUSTER']
X.insert((X.shape[1]),'CLUSTER_5',kmeans)
```

```python
obj = split_data(X.drop(['CPF'],axis=1))
h = obj[0]
df = obj[1]
display(HTML(df.to_html()))
```
